```python
import pandas as pd

df_engelska = pd.read_excel("Data/riket2023_åk9_np.xlsx", sheet_name = "Engelska")
df_engelska = df_engelska.tail(4).reset_index().drop(columns="index")
df_engelska
```

```python
df_engelska.columns
```

```python
df_engelska_clean = df_engelska.rename({"Unnamed: 0": "Plats", 
                    "Unnamed: 1": "Huvudman", 
                    "Unnamed: 2": "Toalt (A-F)", 
                    "Unnamed: 3": "Flickor (A-F)", 
                    "Unnamed: 4": "Pojkar (A-F)",
                    "Unnamed: 5": "Totalt (A-E)", 
                    "Unnamed: 6": "Flickor (A-E)", 
                    "Unnamed: 7": "Pojkar (A-E)", 
                    "Unnamed: 8": "Totalt (poäng)", 
                    "Unnamed: 9": "Flickor (poäng)",
                    "Unnamed: 10": "Pojkar (poäng)"
}, axis=1)
df_engelska_clean
```

```python
df_engelska_clean["Toalt (A-F)"] = df_engelska_clean["Toalt (A-F)"].astype(int)
df_engelska_clean["Totalt (A-E)"] = df_engelska_clean["Totalt (A-E)"].astype(float)
df_engelska_clean
```

```python
totalt_godkända = (df_engelska_clean["Totalt (A-E)"] / 100) * df_engelska_clean["Toalt (A-F)"]
totalt_underkända = df_engelska_clean["Toalt (A-F)"] - totalt_godkända
totalt_underkända
```

```python
df_svenska = pd.read_excel("Data/riket2023_åk9_np.xlsx", sheet_name = "Svenska")
df_svenska = df_svenska.tail(4).reset_index().drop(columns="index")

df_matematik = pd.read_excel("Data/riket2023_åk9_np.xlsx", sheet_name = "Matematik")
df_matematik = df_matematik.tail(4).reset_index().drop(columns="index")
```

```python
df_svenska_clean = df_svenska.rename({"Unnamed: 0": "Plats", 
                    "Unnamed: 1": "Huvudman", 
                    "Unnamed: 2": "Toalt (A-F)", 
                    "Unnamed: 3": "Flickor (A-F)", 
                    "Unnamed: 4": "Pojkar (A-F)",
                    "Unnamed: 5": "Totalt (A-E)", 
                    "Unnamed: 6": "Flickor (A-E)", 
                    "Unnamed: 7": "Pojkar (A-E)", 
                    "Unnamed: 8": "Totalt (poäng)", 
                    "Unnamed: 9": "Flickor (poäng)",
                    "Unnamed: 10": "Pojkar (poäng)"
}, axis=1)
df_matematik_clean = df_matematik.rename({"Unnamed: 0": "Plats", 
                    "Unnamed: 1": "Huvudman", 
                    "Unnamed: 2": "Toalt (A-F)", 
                    "Unnamed: 3": "Flickor (A-F)", 
                    "Unnamed: 4": "Pojkar (A-F)",
                    "Unnamed: 5": "Totalt (A-E)", 
                    "Unnamed: 6": "Flickor (A-E)", 
                    "Unnamed: 7": "Pojkar (A-E)", 
                    "Unnamed: 8": "Totalt (poäng)", 
                    "Unnamed: 9": "Flickor (poäng)",
                    "Unnamed: 10": "Pojkar (poäng)"
}, axis=1)

df_matematik_clean
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

betyg_poäng = {
        "Svenska": df_svenska_clean.query("Huvudman == 'Samtliga'")["Totalt (poäng)"],
        "Engelska": df_engelska_clean.query("Huvudman == 'Samtliga'")["Totalt (poäng)"],
        "Matematik": df_matematik_clean.query("Huvudman == 'Samtliga'")["Totalt (poäng)"]

}

bar = sns.barplot(data=betyg_poäng)

plt.xlabel("Subject")
plt.ylabel("Scorew")

plt.show()
```
