```python
import pandas as pd
df_os = pd.read_csv("Data/athlete_events.csv")

df_os
```

```python
import duckdb as db

db.query("SELECT * FROM df_os").df()
```

```python
df_os.columns
```

```python
db.query("SELECT Name, Age, Sport FROM df_os ORDER BY Age DESC LIMIT 10").df()
```

```python
db.query("SELECT Name, Age, Sport FROM df_os ORDER BY Age LIMIT 10").df()
```

```python
db.query("SELECT Sport, Median(Age) as age_median FROM df_os GROUP BY Sport ORDER BY Age_median DESC LIMIT 5").df()
```

```python
db.query("SELECT Sport, Median(Age) as age_median FROM df_os GROUP BY Sport ORDER BY Age_median LIMIT 5").df()
```

```python
db.query("SELECT NOC, count(Medal) AS medal_sum FROM df_os WHERE Medal = 'Gold' GROUP BY NOC ORDER BY medal_sum DESC LIMIT 10").df()
```

```python

db.query("SELECT NOC, count(Medal) AS medal_sum FROM df_os GROUP BY NOC ORDER BY medal_sum DESC LIMIT 10").df()

```

```python
import seaborn as sns
male_female = db.query("SELECT Year, sex, count(Sex) AS total FROM df_os GROUP BY Year, Sex ORDER BY Year").df()
sns.lineplot(male_female, x="Year", y="total", hue="Sex")
```
